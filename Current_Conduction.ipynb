Electricity conduction in a copper wire is a well-understood process involving the movement of electrons through the metal lattice. Here’s a detailed description of how this conduction occurs:
### Atomic Structure of Argon
The electron configuration for Argon (Ar), which has an atomic number of 18, is:

$$1s^2 \, 2s^2 \, 2p^6 \, 3s^2 \, 3p^6$$

This configuration shows that Argon has a full outer shell of electrons, which makes it chemically inert and a noble gas.

the electrons in the $3s^2$ and $3p^6$ orbitals are in the outermost shell, which is the third electron shell.

### Atomic Structure of Copper

Copper is a metal with an atomic number of 29. Its atomic structure consists of a nucleus surrounded by electrons arranged in electron shells. In the outermost shell (the valence shell), copper has one electron in the $4s$ orbital. The electron configuration for copper is $[Ar] 3d^{10} 4s^1$. This single electron in the $4s$ orbital is loosely bound to the nucleus, which makes it relatively free to move compared to electrons in inner shells.

# Orbital Stuff and Copper

Here’s a detailed description of each type of orbital mentioned in the electron configuration of copper:

### **1. s Orbitals:**

- **1s Orbital:**
  - **Location:** Closest to the nucleus, in the first electron shell.
  - **Shape:** Spherical.
  - **Capacity:** Holds up to 2 electrons.
  - **Description:** This orbital is the lowest energy level and is completely filled in copper.

- **2s Orbital:**
  - **Location:** In the second electron shell, just outside the 1s orbital.
  - **Shape:** Spherical.
  - **Capacity:** Holds up to 2 electrons.
  - **Description:** This orbital is also completely filled in copper.

- **3s Orbital:**
  - **Location:** In the third electron shell, beneath the 3p orbitals.
  - **Shape:** Spherical.
  - **Capacity:** Holds up to 2 electrons.
  - **Description:** This orbital is fully occupied in copper.

- **4s Orbital:**
  - **Location:** In the fourth electron shell, above the 3d orbitals.
  - **Shape:** Spherical.
  - **Capacity:** Holds up to 2 electrons.
  - **Description:** In copper, the 4s orbital has 1 electron due to electron configuration stability considerations.

### **2. p Orbitals:**

- **2p Orbitals:**
  - **Location:** In the second electron shell, above the 2s orbital.
  - **Shape:** Dumbbell-shaped and oriented along the x, y, and z axes (px, py, and pz).
  - **Capacity:** Each p orbital holds up to 2 electrons, with three p orbitals (totaling 6 electrons).
  - **Description:** In copper, the 2p orbitals are fully occupied.

- **3p Orbitals:**
  - **Location:** In the third electron shell, above the 3s orbital.
  - **Shape:** Dumbbell-shaped and oriented along the x, y, and z axes (px, py, and pz).
  - **Capacity:** Each p orbital holds up to 2 electrons, with three p orbitals (totaling 6 electrons).
  - **Description:** These orbitals are fully occupied in copper.

### **3. d Orbitals:**

- **3d Orbitals:**
  - **Location:** In the third electron shell, but higher in energy than the 3s and 3p orbitals. There are five 3d orbitals.
  - **Shape:** More complex shapes (cloverleaf or double-dumbbell).
  - **Capacity:** Each d orbital can hold up to 2 electrons, with a total capacity of 10 electrons across the five d orbitals.
  - **Description:** In copper, the 3d orbitals are fully filled with 10 electrons, contributing to the stability of the atom.

### **Summary:**

- **s Orbitals** are spherical and can hold up to 2 electrons.
- **p Orbitals** are dumbbell-shaped and can hold up to 6 electrons in three separate orbitals.
- **d Orbitals** have more complex shapes and can hold up to 10 electrons in five separate orbitals.

In copper, the arrangement and occupancy of these orbitals are crucial for understanding its chemical behavior and properties.

### Metal Lattice and Electron Sea Model

Copper atoms form a face-centered cubic (FCC) lattice structure in the solid state. In this structure, copper atoms are arranged in a regular pattern, and the metal exhibits high electrical conductivity due to the following characteristics:

1. **Delocalized Electrons**: The outer 4s electrons of copper atoms become delocalized, meaning they are not bound to any specific atom and can move freely throughout the metal. This collection of free-moving electrons forms what is known as an "electron sea" or "cloud" around the positively charged copper ions (which are the copper atoms after losing their valence electrons).

2. **Metallic Bonding**: The metallic bond in copper is the result of electrostatic attractions between the positively charged copper ions and the delocalized electrons. This bonding contributes to the metal's strength and electrical conductivity. The free electrons are highly mobile and can flow through the lattice when an electric field is applied.

### Conduction Mechanism

When an electric potential (voltage) is applied across a copper wire, the following happens:

1. **Electric Field Application**: The voltage creates an electric field in the wire. This electric field exerts a force on the free electrons in the electron sea.

2. **Electron Movement**: The free electrons begin to drift in the direction opposite to the applied electric field (because electrons are negatively charged). This drift is superimposed on the random thermal motion of the electrons.

3. **Current Flow**: The drift of electrons constitutes an electric current. The rate of electron drift (or drift velocity) is directly proportional to the applied electric field, according to Ohm's Law $I = \frac{V}{R}$, where $I$ is the current, $V$ is the voltage, and $R$ is the resistance of the wire.

### Collisions and Resistance

As electrons move through the metal lattice, they frequently collide with the positively charged copper ions and with impurities or defects in the metal. These collisions cause resistance, which impedes the flow of electrons to some extent. The amount of resistance depends on factors such as:

1. **Temperature**: As temperature increases, the lattice ions vibrate more, leading to increased collisions and thus higher resistance.

2. **Purity of Copper**: Impurities and defects in the copper lattice can increase the number of collisions and thus the resistance.

3. **Wire Dimensions**: A longer wire or a wire with a smaller cross-sectional area increases resistance.


The relationship between voltage and electric field is fundamental to understanding how electricity behaves in conductors like copper wires. Here’s a detailed verification of this relationship:

### Voltage and Electric Field Relationship

1. **Electric Potential (Voltage)**: Voltage, or electric potential difference, is defined as the work done to move a unit positive charge from one point to another in an electric field. Mathematically, it's expressed as $V = \frac{W}{Q}$, where $V$ is the voltage, $W$ is the work done, and $Q$ is the charge.

2. **Electric Field Definition**: The electric field $\mathbf{E}$ is defined as the force per unit charge experienced by a positive test charge placed in the field. It is given by $mathbf{E} = \frac{F}{Q}$, where $F$ is the force and $Q$is the charge.

### Relationship Between Voltage and Electric Field

The electric field and voltage are closely related and can be connected by the gradient of the electric potential. In a uniform field, the relationship is:

$$\mathbf{E} = -\frac{dV}{dx}$$

where $$\frac{dV}{dx}$$ represents the spatial rate of change of the electric potential (voltage). 

### Verification

1. **Experimental Basis**: In practice, when a voltage is applied across a conductor, an electric field is indeed established within the conductor. For example, when a battery or power supply creates a potential difference across a copper wire, this difference creates an electric field inside the wire. This electric field drives the flow of free electrons, resulting in an electric current.

2. **Theory**: According to electrostatics, a voltage difference between two points creates an electric field in the region between those points. The electric field $\mathbf{E}$ is the negative gradient of the electric potential $V$, as expressed by the relation $\mathbf{E} = -\nabla V$. This implies that an electric field exists in the region where there is a potential difference.

3. **Ohm’s Law**: Ohm's Law $V = IR$ relates voltage $V$, current $I$, and resistance $R$ in a conductor. This relationship implicitly assumes the presence of an electric field within the conductor, as the current is driven by this field.

The speed at which electrons move through a conductive wire can be described in terms of two different types of velocities: **drift velocity** and **thermal velocity**. These velocities represent different aspects of electron motion in a conductor.

### Drift Velocity

**Drift velocity** is the average velocity of electrons in the direction of the electric field when a current flows through a conductor. It is much smaller than the thermal velocity of electrons and is directly related to the current and electric field within the wire. 

#### Calculation

The drift velocity $v_d$ can be calculated using the formula:

$$v_d = \frac{I}{n e A}$$

where:
- $$ I $$ is the current flowing through the wire.
- $$ n $$ is the number density of charge carriers (electrons) in the wire.
- $$ e $$ is the elementary charge (approximately $ 1.602 \times 10^{-19} $ coulombs).
- $$ A $$ is the cross-sectional area of the wire.

#### Typical Values

For a typical copper wire carrying a current of 1 ampere, the drift velocity is on the order of millimeters per second. For example:

- In a copper wire with a cross-sectional area of 1 mm² (which is $ 1 \times 10^{-6} $ m²) and a current of 1 ampere, the drift velocity of electrons is approximately $ 1.6 \times 10^{-4} $ meters per second (0.16 mm/s).

This low drift velocity is due to the fact that although electrons move rapidly due to thermal motion, the net movement of electrons (drift) in the direction of the electric field is much slower.

### Thermal Velocity

**Thermal velocity** refers to the random motion of electrons due to thermal energy. This velocity is much higher than the drift velocity.

#### Typical Values

At room temperature (around 300 K), the thermal velocity of electrons in a conductor can be on the order of $ 10^5 $ to $ 10^6 $ meters per second. For copper, this is approximately:

- About $ 1.5 \times 10^6 $ meters per second (1.5 million meters per second) for electrons due to thermal motion.

### Summary

- **Drift velocity**: The average velocity of electrons in the direction of the electric field is typically in the range of millimeters per second.
- **Thermal velocity**: The random thermal motion of electrons can be on the order of $ 10^5 $ to $ 10^6 $ meters per second.

Drift velocity is important for understanding how current flows through a conductor, while thermal velocity is significant for understanding the random motion of electrons due to thermal effects.

# Intrinsic Silicon

Intrinsic silicon, which is pure silicon without any doping, has an electron configuration that reflects its position in the periodic table. Silicon has an atomic number of 14, meaning it has 14 electrons. Its electron configuration is:

$$ 1s^2 \, 2s^2 \, 2p^6 \, 3s^2 \, 3p^2$$

Here's a breakdown of the electron configuration:

- **1s²**: 2 electrons in the first shell (1s orbital)
- **2s²**: 2 electrons in the second shell (2s orbital)
- **2p⁶**: 6 electrons in the second shell (2p orbitals)
- **3s²**: 2 electrons in the third shell (3s orbital)
- **3p²**: 2 electrons in the third shell (3p orbitals)

In its crystalline form, silicon forms a diamond-like lattice structure where each silicon atom shares its four valence electrons with four neighboring silicon atoms, creating strong covalent bonds. This bonding arrangement results in a stable structure but also means that the electrons in the 3p orbitals are effectively used in bonding, contributing to the semiconductor properties of silicon.
